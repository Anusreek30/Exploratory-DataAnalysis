```python
import pandas as pd
df=pd.read_csv("/Flavors.csv")
df
```

```python
#groupby
group_by_frame=df.groupby("Base Flavor")
print(group_by_frame)
```

```python
group_by_frame.mean()
```

```python
numerical_cols = df.select_dtypes(include='number').columns
df.groupby('Base Flavor')[numerical_cols].mean()
```

```python
#count
df.groupby('Base Flavor')[numerical_cols].count()
```

```python
#sum
df.groupby('Base Flavor')[numerical_cols].sum()
```

```python
#using agg function to find particular value of a column
df.groupby('Base Flavor').agg({"Flavor Rating":["max","min","mean"],"Texture Rating":["max","min","sum"]})
```

```python
#grouping two columns
df.groupby(['Base Flavor','Liked']).agg({"Texture Rating":["max","min"]})
```
